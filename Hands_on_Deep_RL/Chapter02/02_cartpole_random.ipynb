```python
import gym
```

```python
env = gym.make("CartPole-v0")

total_reward = 0.0
total_steps = 0
obs = env.reset()
```

```python
while True:
    action = env.action_space.sample()
    obs, reward, done, _ = env.step(action)
    total_reward += reward
    total_steps += 1
    print(f'reward: {reward}, total steps: {total_steps}')
    if done:
        break

    print("Episode done in %d steps, total reward %.2f" % (
        total_steps, total_reward))
```
